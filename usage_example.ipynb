```python
import numpy as np
import pandas as pd
from src.pipeline import train_model
```

### Build toy dataset for example

```python
#-- Parameters for dataset generation
n_clusters = 3
cluster_sizes = 900 // n_clusters

#-- Generate dataset
toy_df = []

for cluster in range(n_clusters):
    cluster_df = pd.DataFrame({
        'cat_var_1': np.random.choice(['expert', 'norm_user', 'new_user'], size=cluster_sizes),
        'cat_var_2': np.random.choice(['high_income', 'meidum_income', 'low_income'], size=cluster_sizes),
        'num1': cluster * 50 + np.random.normal(scale=5, size=cluster_sizes),
        'num2': cluster * 20 + np.random.normal(scale=3, size=cluster_sizes),
    })
    toy_df.append(cluster_df)

df = pd.concat(toy_df, ignore_index=True)
df.head(10)
```

### Run model:
1) Define numerical variables
2) Define categorical variables
3) Run model

```python
#-- Define features
numerical_columns   = ['num1', 'num2']
categorical_columns = ['cat_var_1', 'cat_var_2']

#-- Train model
results_df = train_model(data = df, k=3, categorical_cols=categorical_columns, numerical_cols=numerical_columns, batch_size=100)

results_df
```
